```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bank_Churn.csv")
df.head()
```

```python
df.shape
```

```python
df['Gender'].value_counts()
```

```python
df['Exited'].value_counts()
```

```python
X = df.iloc[:, 3:13]
y = df['Exited']
```

```python
X.head()
```

```python
y.head()
```

```python
# Creating Dummy variables
gender = pd.get_dummies(X['Gender'])
geography = pd.get_dummies(X['Geography'])
```

```python
#Concatenating them back to the dataset
X = pd.concat([X,geography,gender], axis=1)
X.head()
```

```python
# Removing the Features not required
X = X.drop(['Geography', 'Gender'], axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

## Creating an ANN 

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Initializing the Network
ann = Sequential()
```

```python
# Input Layer
ann.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu', input_dim=13))
ann.add(Dropout(0.3))

#Second Layer
ann.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu'))
ann.add(Dropout(0.4))
```

```python
#Creating an output layer
ann.add(Dense(units=1, kernel_initializer='glorot_uniform' , activation='sigmoid'))
```

```python
ann.summary()
```

```python
#Compiling the ANN
ann.compile(optimizer='adam', loss='binary_crossentropy' , metrics=['accuracy'])
```

```python
# Fitting the Data
model = ann.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs=100)
```

```python
y_pred = ann.predict(X_test)
y_pred = y_pred > 0.5
y_pred[:5]
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#Claclulating accuracy score
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
score
```
